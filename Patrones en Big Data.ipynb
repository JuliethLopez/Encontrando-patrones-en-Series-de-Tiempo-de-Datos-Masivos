## Definición de big data
Los origenes de Big Data vienen de tiempo atras, cada avance tecnologico y cada idea fueron pequeños pasos que llevaron a lo que conocemos hoy, desde la memoria virtual, el uso del internet hasta el internet de las cosas, sin embargo, el concepto de big data cobró impulso a principios de la década de 2000 cuando el analista de la industria, Doug Laney, articuló la definición actual de grandes datos como las tres V's:

- Volumen.
- Velocidad: Dado el uso de la nueva tecnologia surge la necesidad de manejar estos grandes volumenes de datos en tiempo casi real.
- Variedad: Los datos pueden estar en distintos formatos, desde datos numéricos estructurados en bases de datos tradicionales hasta documentos de texto no estructurados, correos electrónicos, vídeos, audios, datos de teletipo y transacciones financieras.

Algunos consideran que deben ser 5 V's, las otras dos dimensiones son:

- Variabilidad: Los flujos de datos son impredecibles, es decir cambian a menudo, esto puede indicar las modas en los medios sociales, y exige una gestion de los picos de carga de datos diarios, estacionales o por eventos.

- Veracidad: Debido a que los datos provienen de tantas fuentes diferentes, es difícil correlacionar, jerarquizar, comparar, limpiar y transformar los datos a través de los sistemas. Si esto no se realiza adecuadamente, los datos pueden salirse de control.

De forma general Big data aplica cuando un conjunto de datos es tan grande, rápido o complejo que es dificil o imposible procesarlos con métodos tradicionales o con la tecnología y software tradicional, es decir, cuando los datos no se pueden procesar en un computador personal, es decir, es tan grande que se debe almacenar en mas de un lugar al tiempo, o no se pueden obtener de un solo lugar.

¿Qué ventaja tiene poder analizar millones de datos de distintas procedencias?Permite que la toma de decisiones sea mucho más rápida, precisa y efectiva. Dados los avances tecnologicos y el incremento de aparatos tecnologicos capaces de recolectar información, cada día es mas facil recopilar información y tambien guardarla, como dato curioso, la capacidad de computación de un iPhone ahora es mayor que la que tenía la NASA cuando envío el Apolo 11 a la Luna, la capacidad de computación se refiere a la velocidad de procesamiento,capacidad de almacenamiento y capacidad de memoria.

## Explicación del método
El problema esencial en el modelamiento de las series de tiempo es poder identificar el comportamiento de cualquier variable aleatoria en un modelo con un periodo de tiempo largo o corto cuando los datos son masivos. Para series de tiempo de datos masivos es **imposible asumir estacionariedad**, un posible acercamiento es separar la serie en subseries o como es llamado en el contexto de datos masivos "regimenes", para ajustar en un principio un modelo a cada régimen. Antes de empezar a estudiar la serie de tiempo es importante encontrar los puntos de cambio que son los que determinan los puntos de inicio y de finlización de los regimenes. Los modelos ajustados a los regimenes explican algunas carácteristicas de media o de escala e incluso puede especificar algunas caracteristicas dependientes del tiempo como la tendencia o la autocorrelación.

Para empezar es necesario hallar los puntos de cambio, o inicio y final, de cada régimen y más adelante se podría estar interesado en la longitud de cada régimen. La complejidad de hallar los puntos de cambio radica en la suavidad de los modelos individuales. Un secuencia particular de modelos individuales o modelos locales define un patron y se denota por 
$$
P_{ri} = (\mu_i(t),\mu_{i+1}(t),...,\mu_{i+r}(t))
$$
donde $\mu_i$ es un vector cuyos elementos cuantifican todos los aspectos importantes del modelo. Mientras que un modelo es una función de tiempo con descripciones de  las componentes de otros modelos (similar a una función a trozos).

Los principales temas que se buscan resolver son el aprendizaje en las series de tiempo, que generalmente se refiere al agrupamiento o clasificación de series y la predicción o el forecasting, sin embargo los autores ven el problema de predicción como un problema de clasificación porque se busca un clasificador basado en patrones, la métrica que se utiliza es $(P_{ri}, P_{sj})$, donde $P_{ri}$ es un patrón que consiste en la secuencia $\mu_i(t),\mu_{i+1}(t),...,\mu_{i+r}(t)$ y $P_{sj}$ es un patrón que consiste en una secuencia sobre $s$ regimenes empezando con el j-esimo.

Por lo general, el analisis de datos masivos requiere una reducción de datos. 

### Puntos de cambio
Cuando se tienen muchos datos es posible desarrollar modelos muy buenos pero muy complejos, sin embargo esto puede llegar a causar sobreajuste, lo que se conoce como "overfitting". Una de las principales técnicas para encontrar los puntos de cambio es el **suavizamiento de tendencias alternadas** (alternative trends smoothing).

##Clasificación y predicción

Hay un interés considerable actualmente en el aprenizaje de series de tiempo. "Aprendizaje" generalmente significa clustering y/o clasificación. Esta es una de las principales motivaciones  en el reconocimiento de patrones en series de tiempo.
La predicción es otra de las motivaciones, el análisis de tendencia simple puede realizarse mediante modelos como ARMA u otras técnicas de análisis. El problema de predicción es visto como un problema simple de clasificación en el cual el aprendizaje es usado para desarrollar un clasificador basado en patrones.

##Medidas de similaridad/disimilaridad 

El clustering o clasificación es a menudo basado en medidas de disimilaridad de eventos en un conjunto. Para clustering y clasificación de una serie de tiempo se necesita una métrica $\rho (P_{ri}, P_{sj})$, donde $P_{ri}$ es un patrón que consiste en una secuencia $\mu_{i},\mu_{i+i},...,\mu_{i+r}$ y $P_{sj}$ es un patrón que consiste en la secuencia de s regimenes empezando en el j-ésimo.

##Reducción de datos y puntos de cambio 

En el análisis de datos masivos, sean series de tiempo o no, usualmente se comienza con al guna forma de reducción de datos. Esto involucra el cálculo de estadísticas resumen que miden tendencias centrales y propagación o dispersión. Sin embargo, en el caso de datos masivos, así se asuma un modelo para todos los datos puede llegar a perderse información, además de estadísticas que pueden no ser robustas, en este caso se tiene que hacer un análisis de posibles datos outliers. En este contexto se pueden encontrar los puntos de cambio como los puntos donde el comportamiento de la propiedad de interés cambia. Los puntos de cambio solo tienen interpretación en el contexto de un modelo. El modelo para los datos observados puede ser uno parámetrico, como un modelo ARMA, o uno no parámetrico, como el de medianas constantes. 

Desde cierta perspeciva, el problema de indentificación de los puntos de cambio puede ser visto como parte del proceso de construcción de un modelo (con fuertes restricciones). A continuación el enfoque es la identificación de los puntos de cambio en un modelo lineal simple por partes de una varible aleatoria observable. No se asumirán momentos finitos, así que al parámetro de tendencia central se le llamará la "mediana" y al de variabilidad "escala".

## Modelos constantes por partes

El modelo más simple para series de tiempo con puntos de cambio es el que cada régimen es modelado por una constante. Esta constante es algún valor promedio de los datos en cada régimen. Para nuestro propósito, la naturaleza de ese "promedio" no es relevante; sin embargo debido a la posibilidad de colas pesadas en las distribuciones de frecuencia y de asimetría de los datos, se pensará en ese promedio como una mediana. 
Hay varios enfoques para modelar series de tiempo con valores de la mediana que cambian a través del tiempo. El primer paso es determinar los breakpoints o escoger una aproximación de los breakpoints de los datos con un modelo por partes aunque el modelo no esté generado de este.

##Modelos con cambios de escala
Existen modelos en los que la variabilidad del proceso no se mantiene a través del tiempo, esto quiere decir que que cambia de escala o es volátil. La volatilidad en clustering es un ejemplo de cambios en una serie de tiempo en la cual la localización puede estar relativamente sin cambios, pero la escala cambia de un régimen a otro. Con el objetivo de identificar cambios en la escala se debe contar con alguna medida de volatilidad. Medir la volatilidad en series de tiempo no será intuitivo y menos cuando esta cambia.

## Tendencias

El principal interés en el análisis de patrones en series de tiempo se enfoca en el cambio en la tendencia. Los modelos lineales simples de series de tiempo son aquellos en donde la tendencia crece o decrece y los puntos de cambio son aquellos en donde esta cambia. Los gráficos de punto y figura son buenos para la identificación de puntos de cambio y la cantidad de cambio se mide con una tendencia más arriba o más abajo con respecto a la anterior. 
Algunos breakpoints entre tendencia son más interesantes que otros, dependiendo en la extensión del cambio de la tendencia. Dentro de un regimen el cual solo una tendencia es dominante tendencia menores de diferente dirección o diferente magnitud pueden ocurrir. Esto aumenta el problema de regímenes adicionales, posiblemente cayendo en overfitting o en dejar un régimen en el cual muchos puntos se desvían significativamente del modelo ajustado.

#Construcción del modelo

Uno de los principales objetivos de construir un modelo para una serie de tiempo es reducir la cantidad de datos usando una representación aproximada de los datos. Aunque los modelos estándar como ARIMA o GARCH podrían entender el proceso de generación de los datos y también proveen una aproximación o suavizamiento de estos y así una importante reducción de los datos, se propone una aproximación basada en los modelos simples por partes. Estos dependen de la identificación de puntos de cambio y el modelamiento dentro de cada régimen individual.  
Debido a que la identificación de los puntos de cambio, esto es, la identificación de los regímenes, está altamente ligada al ajuste de los modelos dentro de los regímenes individuales, no es posible separar estos dos pasos. Usualmente, una forma del modelo es escogida y luego los regímenes son escogidos basados en qué tan bueno es el ajuste del potencial modelo.

##Métodos de lotes

Para ajustar los modelos constantes por partes, hay varias maneras sencillas de determinar los valores de las constantes de aproximación, Si todos los datos están disponibles un proceso por lotes simple es usar los cuantiles muestrales de los datos, que corresponde a un modelo paramétrico, y luego identificar los regímenes como los grupos (clusters) alrededor de esos cuantiles. 
Otro enfoque de lotes (simple) es ajustar un solo modelo y luego identificar subsecuencias basadas en puntos ooutliers con respecto al modelo ajustado. Este proceso es repetido recursivamente en subsecuencias, empezando con la secuencia entera.
Dada una tendencia lineal sobre algún régimen, se definen los puntos "perceptualmente importantes" como aquellos candidatos a ser puntos de cambio. Estos puntos son los que se desvían de la línea de tenencia.

##Métodos en línea

Los métodos de lotes, como los que basan los modelos locales en cuantiles muestrales de toda la serie de tiempo, o aquellos que identifican subsecuencias recursivamente con modelos locales tienen limitada aplicabilidad. En muchas apliaciones del análisis de series de tiempo, nuevos daton son continuamente adquiridos, así, un método en línea es preferible respecto a uno de lotes. Un método en línea accede a los datos, una observación al tiempo y puede retener solo una cantidad predeterminada de datos para usar en cálculos postriores.

#Construcción del modelo: Suavizamiento de tendencias alternantes

Un método para identificar puntos de cambio en una serie de tiempo basada en tendencias alternantes, llamada "alternating trends smoothing", o ATS, es dada en el algoritmo 1. Depende de un parámetro  de suavizamiento $h$, el cual especifica el tamaño del paso dentro de los cuales se buscan los puntos de cambio.

**Algoritmo 1**: $ATS(h)$

1. Establecer $d=1$ (contador de punto de cambio) 

2. While(más datos en el primer paso del tiempo) 
    
(a) for $i=1,2,..,m$, donde $m=h$ si hay $h$ datos adicionales disponibles en otro caso $m$ es el último dato:

input $x_i$;

(b) Establecer $b_d =1$ y $c_d =x_1$

(c)determinar $j_+$, $j_-$, $x_{j_+}$, $x_{j_-}$ tales que $x_{j_+}=max(x_1, x_2,.., x_h)$ y $x_{j_-}=min(x_1, x_2,.., x_h)$

(d) Establecer $s=(x_k - x_i)/(k-i)$ y $r=sign(s)$

(e) while($r=0$), continuar input: terminar en el último dato.

<br>

3. Establecer $j=i$ (índice del último dato en el paso previo); y establecer $d=d+1$

4. While(más datos)

(a) for $i=j+1, j+2,...,j+m$, donde $m=h$ si hay $h$ datos adicionales, en otro caso $m=j+m$ es el último dato:

input $x_i$;

- **i.** while($sign(s)=r$)

  - A. Establecer $k=min(i+h,n)$

  -  B. if ($k=i$) break

  - C. Establecer $s=(x_k - x_j)/(k-j)$

  - D. Establecer $j=k$

- **ii.** determinar $j_+$ tal que $rx_{j_+}$ es el máximo de $rx_{j+1},...,rx_{j+m}$

- **iii.** Establecer $b_d = j_+$; y $c_d =x_{j_+}$

- **iv.** Establecer $d=d+1$; $j=j_+$; y $r=-r$

(b) Establecer $b_d=j_+$; y $c_d=x_{j_+}$ .

<br>

La salida de este algoritmo aplicado a una serie de tiempo $x_1,x_2,...$ es $(b_1,c_1),(b_2,c_2),...$, donde $b_1=1$, $b_2=t^{(2)}$, y $c_1=x_{t^{(2)}}$, con $t^{(2)}$ el tiempo en el cual el primer cambio de tendencia ocurrió.

Entre dos breakpoints la tendencia está representada por la pendiente de los valores de la serie en dos puntos divididos por el tiempo entre ellos, y la serie suavizada es la línea de tendencia por partes que conecta los valores de los puntos de cambio, esto es conocido como patrón cabeza-y-hombros (head-and-shoulders). El método es efectivo para encontrar patrones. A continuación se muestra un ejemplo

## El parámetro de afinación 

El parámetro de afinación $h$ en el algoritmo 1 es un "step size". El proceso de identificar el siguiente punto de cambio comienza con el dato un $h$ más allá del actual punto de cambio. Valores grande de $h$ tienden a incrementar las distancias entre los puntos de cambio, pero la distancia real entre estos puede ser tan pequeña como una unidad de tiempo.
Aunque en la implementación estándar, la identificación de las tendencias en el ATS está basada en puntos individuales, el comportamiento agregado tiende a dominar, especialmente después de la primera tendencia. En la identificación del punto de cambio al final del primer régimen, la dependencia de la tendencia en el punto un h más allá puede conducir a una identificación errónea de la tendencia. Esto debido a que el ATS trabaja identificando puntos de cambio basado en cambios de la tendencia, y en la primera tendencia no hay una tendencia previa para ser comparadas. Esto puede resultar en una tendencia determinada por puntos $x_1$ y $x_h$ que difieren completamente de la tendencia aparente en los puntos $x_1,...,x_{h-1}$. Por esta razón, una modificación simple del algoritmo del ATS en el primer paso es usar algún otro criterio para determinar la tendencia. Un enfoque es un ajuste de mínimos cuadrados de una línea a través de $x_1$ y que se acerca a los puntos $x_2,...,x_h$. Debido a la restricción de que la línea tiene que atravesar a $x_1$, el criterio de mínimos cuadrados puede ser ponderado por los leverages de los otros puntos. Si se asume que el espaciado temporal es igual sobre los puntos $x_1,...,x_h$ la pendiente de mínimos cuadrados es

$arg min \displaystyle\sum_{i=2}^{h} (x_i-s_i)^2/i= (h^2-h) /\displaystyle\sum_{i=2}^{h} x_i$ 

Si todos los valores d los h primeros datos siguen la misma tendencia, la modificación no tiene efecto. Esta modificación también puede ser en cada paso, y a menudo resulta lo que parece visualmente un mejor ajuste. Pero la modificación solo tiene un efecto importante en el primer régimen. Un $h$ de n/10 trabaja bien, sin embargo, es recomendable experientar con varios valores.

El ATS puede aplicarse iterativamente, es decir, aplicarse a los puntos de cambio encontrados en la primera aplicación, de esta manera se puede obtener un resultado más preciso, sin embargo hay que tener cuidado pues se llega a un modelo con mucho ruido y que puede considerarse overfitting. ##Modificaciones y Extensiones

Un ajuste de tendencias alternantes puede ser modificado de varias maneras. Una de ellas es subdividir los regímenes identificando los puntos de cambio dentro de algún régimen basado en la desviación de la línea de tendencia dentro del régimen. Sin embargo, la simplicidad computacional del ATS puede perderse. Este enfoque podría requerir un parámetro de afinación para cuantificar la significancia del cambio de tendencia cuando el signo de la tendencia no ha cambiado. Las alteraciones podrían estar basados en el número de puntos de cambio o en qué tan bueno es el ajuste dentro del régimen. Esta modificación requiere de un parámetro adicional de afinamiento para decidir si el régimen va a ser "separado" en subregímenes. En datos ruidosos o que presenten sobreajuste esta medida no será provechosa. 

##Líneas de confianza

En lo que hemos dicho no se ha asumido ningún modelo de probabilidad específico. Se quiere identificar los límites de confianza dado un régimen que generalmente donde un modelo tiene la misma dirección que la tendencia de esa región. Sin embargo, las líneas de confianza no dependen de ningún modelo. Para una serie de tiempo dada ${x_t : t = 1,…,n}$, buscamos una línea límite inferior, denominada línea de soporte $x = a+bt$ que satisface el problema de optimización
$$
min_{a,b} \sum^n_{i=1} p(x_t – a + bt) , 
$$

con $x_t >= a + bt$, dónde $p(.)$ es no negativa tal que $\sum^n_{i=1} p(.)$ con vector $(.)$ es normal. De forma análoga para la línea superior denominada línea de resistencia, pero con $x_t <= a + bt$.

Esta tarea de optimización puede ser complicada, sin embargo con $p(.)=|.|$ se vuelve más sencillo, esto se resuelve con un método que se basa en dos hechos, el primer hecho es que la norma $L_1$ satisface la desigualdad triangular con igualdad, es decir $||y + z||_1 = ||y||_1 + ||_z||_1$, el segundo hecho es que los datos son igualmente espaciados a través del tiempo. Y aunque la solución no es única no supone un problema.
Para los valores óptimos de $a$ y $b$ hay 3 casos:
-	K=(n+1)/2 (se considera en el paso 3 del algoritmo)
-	K<=n/2 (se considera en el paso 4 del algoritmo)
-	k>=n/2 (se considera en el paso 5 del algoritmo)

##Patrones
Recordemos que hemos propuesto varios modelos para un patrón, el modelo constante por partes $\mu_i (t)=c_i I_{[t_{i1}.t_{i2}]} (t)$ con $I_s (t)$ función indicadora que depende de si t pertenece a S. Y el modelo ATS $\mu_i (t) = (a_i + s_i t) I_{[t_{i1}, t_{i2}]} (t)$ desarrollado en el 1er algoritmo. Para comparar diferentes series de tiempo, comparamos el patrón de la serie con otros patrones de otra series usando la métrica $\rho(P_{ri}, P_{sj})$ donde $P_{ri}$ es un patron que consiste en la secuencia $\mu_i, \mu_{i+1} ,…, \mu_{i+r}$ y $P_{sj}$ es un patron que consiste en una secuncia sobre s régimenes comenzando en el j-esimo.
Lo primero a tener en cuenta es el registro de las diferentes series de tiempo, es decir cambiando y escalando la serie de tiempo para que los regímenes en la serie de tiempo separados puedan ser comparados, también es posible hacer una reducción de los datos.

##Reducción de datos: Representación simbólica

Los componentes individuales del patrón $P_{ri}$ pueden contener detalles del modelo, algunos detalles pueden ser suprimidos, pero las características que destacan deben ser retenidas. Por ejemplo, para una secuencia de modelos constantes, las características más importantes pueden ser una secuencia de indicadores sobre si $c_i$ era de tamaño menor, medio y largo; eso es, el patron es una secuencia de la forma *abcde...*, donde cada *a*, *b* y demás son valores que indican "corto", "medio" y "largo". Ésta representación puede ser util para clasificar, sin embargo, no incluye información sobre $t_{i1}s$ o $t_{i2}s$ o incluso valores exactos de $c_i s$. Este paso en el que clasficica y se representa la informacion con un solo simbolo es utilizado en minerio de datos de series de tiempo. Una aproximación simbolica ampliamente utilizada para series de tiempo es SAX, que se basa en una modificacion de la secuencia de modelos constantes por partes(PAA) y se puede ver en Lin et al (2007). Otro metodo es el NSAR para una representación simbolica no paramétrica y se puede ver en He et al (2014), y se basa en transformacines wavelet.

La transformación de modelos con parámetros cuantitativos a modelos de simbolos categoricos requiere una definición a priori de los símbolos, establecer previamente una lista, o una fórmula para definir nuevos símbolos. En el procesamiento por lotes de los datos, el rengo de posibles modelos puede ser determinado antes de la transformación de modelos cuantitativos en modelos categóricos. En ATS, cada modelo se caracteriza por cuatro números reales y una secuencia de r modelos se caracteriza por $2r + 2$ números reales. Para reducir aún más los datos, los números reales se agrupan en grupos ordenados que se asocian con un conjunto único de símbolos. La sustitución de modelos con parametrizaciones numéricas continuas por representaciones simbólicas resulta en la pérdida de datos.

## Patrones simbólicos de tendencia (STP)

La aproximación simbólica del SAX está basada en un tipo de modelamiento de constantes por partes llamados “aproximación agregada por partes” (PAA). La misma idea del SAX puede ser aplicada a los modelos en el ATS. Estos símbolos consisten en pares de símbolos o silabas formados por una selección de una consonante o vocal con los siguientes significados:

- J Creciente muy corta
- K Creciente corta
- L Creciente media
- M Creciente larga
- N Creciente muy larga
<br>

- P Decreciente muy corta
- Q Decreciente corta
- R Decreciente media
- S Decreciente larga
- T Decreciente muy larga
<br>

- A Magnitud muy corta
- E Magnitud corta
- I Magnitud media
- O Magnitud larga
- U Magnitud muy larga



Que representa la magnitud de una tendencia. En muchos casos, las vocales pueden representar la magnitud del cambio en valor en lugar de la magnitud de la tendencia, eso es la pendiente del segmento entre los puntos de cambio. En cada caso, las letras individuales representan cambios en las magnitudes. Por ejemplo, “P” representa una tendencia decreciente de corta duración, y “A” representa una tendencia de magnitud muy pequeña.
Si estos símbolos son definidos y asignados en un lote, ellos pueden representar cuantiles (o cuantiles aproximados)de la serie de tiempo completamente observada. Por ejemplo, para la gráfica en la sección de Suavizamiento de tendencia alternantes, se tienen que la representación simbólica es

## Patrones en líneas 

Después de hacer cualquier tipo de reducción de datos, hay mejores oportunidades para identificar patrones. Las tendencias son una forma simple de reducción de datos que ofrecen varios métodos de identificación de patrones. Las líneas limites discutidas anteriormente tienen los mismos tipos de características, desniveles y duración. Cuando las lineas son determinadas en regiones determinadas por el ATS, sus pendientes normalmente tienen el mismo signo que sus tendencias. Otra característica interesante de las líneas delimitadas es su relación con cada otra; en particular, si parecen converger o divergir. En la figura 6.10, por ejemplo, podemos ver que las líneas delimitadoras en el régimen más a la izquierda parecen estar convergiendo mientras que los del segundo régimen desde la izquierda parecen diverger. Los analistas de seguridad técnica a veces dan significado a tales patrones.

## Clustering y clasificación de series de tiempo

En clasificación o agrupamiento de los datos se necesitan medidas apropiadas de similaridad o no similaridad. Las métricas que se suelen usar son las inducidas por la norma. Para una clase dada de patrones, ya sea que estén definidos en un conjunto finito de símbolos o en Rd. Para métricas inducidas por normas, la relación de equivalencia entre cualquier conjunto de normas produce una equivalencia de métrica. Esta equivalencia se traslada a métricas en un conjunto de símbolos ordenados. A pesar de la equivlencia de metricas, diferentes metricas conducen a distintas clasificaciones. Los principales problemas son los de scaling y "junk". Las 3 figuras mostradas en ña figura 6.11 pueden estar asociadas por las aproximaciones STP

$$
JO, PU, ME, RA\\
MO, RU, KE, PA\\
MO, PI, JA, RI, KE, PA\\
$$

Para reconocer la similaridad de patrones se puede empezar por una reduccion de datos para lidiar con problemas de escala en los tiempos, por ejmplo usando solo arriba y abajo, los primeros dos patrones son el mismo:

$$
+O, -U, +E, -A
$$

Un modelo de corta duración y cambios pequeños es un candidato para modos superfluos; es decir, uno que se pueda suavizar mediante combinaciones
con modelos cercanos. Aplicando este enfoque a la serie de tiempo suavizada en el gráfico derecho de la figura 6.11 resultaria en la combinación del segundo al cuarto modelo en un solo modelo, que seria representado como RU o -U.


Este enfoque del problema implica combinaciones y ajustes para todos los modelos en un conjunto de patrones, por ello no es del todo satisfactorio.
Para el agrupamiento y la clasificación no necesitamos que los patrones sean exactamente iguales, por lo que otro enfoque se basaría en el uso de la métrica adecuada.

Una métrica que pondera las diferencias en la dirección de una tendencia mucho más que diferencias en la longitud de dos tendencias en la misma dirección lograrían algunos de los mismo efecto que considerar la duración como una variable binaria. La clasificación de series de tiempo está estrechamente relacionada con el problema estándar de predicción o pronóstico en series de tiempo. Para un patrón dado, el valor predicho es simplemente la clase predicha del patrón.

```python
def ATS(x, h):
    n=len(x)
    x=x.values
    sp=0
    changepoints=[x[0][0]]
    cplog=[0]
    staring_x=sp
    b=6*(np.sum(np.array(x[1:h+1])*np.array([i for i in range(1,h+1)]))-x[0]*h*(h+1)/2)/(h*(h+1)*(2*h+1))
    b=b[0]
    ep=sp+h
    diff=x[ep]-x[sp]
    while np.sign(diff)!=np.sign(b):
        ep-=1
        if ep<=1:
            print("datos constantes en el primer paso")
            return()
    while diff==0:
        ep+=1
        if ep>=n-1:
            print("pendiente 0 y datos constantes después del primer paso")
            return()
        diff=x[ep]-x[sp]
    pendiente=(x[ep]-x[sp])/h
    cs=np.sign(pendiente)
    while ep<n-1:
        spstart=sp
        while np.sign(pendiente)==cs:
            ep=min(sp+h, n-1)
            if sp==ep: 
                break
            diff=x[ep]-x[sp]
            while diff==0 & ep>sp+1:
                ep-=1
                diff=x[ep]-x[sp]
            if diff==0:
                ep=min(sp+h, n-1)
                while diff==0 & ep<n:
                    ep+=1
                    diff=x[ep]-x[sp]
            pendiente=diff/(ep-sp)
            sp=ep
        sp=spstart-1+list(reversed(np.where(cs*x[spstart:sp]==max(cs*x[spstart:sp]))[0]))[0]
        changepoints.append(x[sp][0])
        cplog.append(sp)
        staring_x=sp
        cs=-cs
    changepoints.append(x[n-1][0])
    cplog.append(n-1)
    return(cplog,changepoints)
```

```python
from sklearn.linear_model import Lasso

def BoundingLines (x, env):
    x=x.values
    n=len(x)
    
    ab=Lasso(alpha=1.e-7).fit(np.array([i for i in range(1,n+1)]).reshape(-1,1), x)
    a=ab.intercept_
    bc=ab.coef_
    r=np.array(x)-a-bc*np.array([i for i in range(1,n+1)])
    
    maxr=max(r*env)
    
    k=np.where(r*env==maxr)[0][0]
    dc=sum(abs(r-env*maxr))
    if (n/2==(n-1)/2) & (k==n/2):
        x1=maxr+a+bc
        x2=maxr+a+bc*n
        #print("problem in one")
        return([x1[0], x2[0]])
    if k<=n/2:
        tans=env*(x[k+1:n]-x[k])/np.array([i for i in range(1, n-k)])
        maxtanright=max(tans)
        j=np.where(tans==maxtanright)[0]+k
        bright=(x[k]-x[j])/(k-j)
        x1=x[k]-bright*(k-1)
        x2=x[k]+bright*(n-k)
        #print("problem in two")
        return(x1[0], x2[0])
    if k>n/2:
        tans = env*(x[0:k]-x[k])/list(reversed([i for i in range(1,k+1)]))
        maxtanleft=max(tans)
        i=np.where(tans==maxtanleft)[0]
        bleft=(x[k]-x[i])/(k-i)
        x1=x[k]-bleft*(k-1)
        x2=x[k]+bleft*(n-k)
        #print("problem in three")
        return([x1[0], x2[0]])
```
